```python
import pandas as pd 

# reading dataset 
df = pd.read_csv( 'casual_data_windows.csv' )

df.head(10)
```

```python
conv_dataset = [] 

for k, row in df.iterrows(): 
    conv = []
    conv.append( row['0'])
    conv.append( row['1'])
    conv.append( row['2'])
    
    conv_dataset.append( conv )

    if k > 10: 
        break

conv_dataset 
```

```python
from chatterbot.trainers import ListTrainer, ChatterBotCorpusTrainer
from chatterbot import ChatBot
from os.path import exists

#  creating instance of chatbot 
chatbot = ChatBot( 
    'John Wick',
    logic_adapters=[
        'chatterbot.logic.MathematicalEvaluation',
        'chatterbot.logic.TimeLogicAdapter'
    ],
)



if not exists( 'db.sqlite3' ):
    trainer = ListTrainer( chatbot )

    # training using the reddit dataset
    for conversation in conv_dataset: 
        trainer.train( conversation )    

while True: 
    # inputting the request statement
    request = input("You : ")

    # Get a response to an input statement
    print( chatbot.name, ": ",chatbot.get_response(request), sep=' ' ) 
```
